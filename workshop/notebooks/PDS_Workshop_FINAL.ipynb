### Visual literacy (Trumbo, 1999)


- Visual thinking (how do we organise and perceive images via their attributes)
  - eyes + brain; mental imaging
  - creative and intuitive thinking; pattern recognition

- Visual learning (developing visual images for intended purpose)
  - representing ideas and concepts through visuals
  - craft of visualisation 




- Visual communication (using visuals to express ideas and convey meanings)
  - design
  - audience

### Persuasive communication (Brunett et al., 2019)


- Central route - thoughtful consideration; more congnitive energy; lasting formation of attitidues, preferences
- Peripherial route - affective association; less congnitive energy; short-term formation of attitudes, preferences
  * Message of (interactive) visuals &rarr; to engage individuals who are least likely to engage
  
> <div class="alert alert-block alert-info"> 
<b>RECALL</b> Aristotle's presuasion criteria targeted at insight (ethos, logos, pathos, telos, kairos)  
</div>
 
 

### Exploratory vs explanatory visualisation (Echeverria et al., 2018; Dzuranin, 2022)


- Exploratory: 
  - exploring/searching data to find insights (e.g., visuals for descriptive statistics)
  - better underlying understanding of the data
 
> <div class="alert alert-block alert-warning"> 
 Link to exploratory (information) vs explanatory (communication) approaches to data communication
 </div>



- Explanatory:
  - **TO COMMUNICATE and PRESENT** the insight (results; findings) in clear and concise manner
  - inspiring the call for action
  - car dashboard example
  
> <div class="alert alert-block alert-info"> 
<b>RECALL</b> Data storytelling Venn diagram
</div>

### Visualisation ecosystems 
Kosara et al. (2016)



#### Visualisation ecosystems - MEANS (providing value of data)

- Analysis 
  - domain experts designing visuals for their own analysis; exploration
  > ***deriving*** insight


- Guided analysis
  - leading the user in exploration (e.g., comparing smartphone features)
- Presentation and communication 
  - persuasive and memorable information &rarr; call to action 
  > ***providing*** insight 

#### Visualisation ecosystems - design priorities across MEANS 

- Analysis (exploration)
  - generality
  - quick iteration - aksing different question to your data through visuals
  - dead ends - adding irrelevant information which makes it harder to follow the main idea
  - trust
  - fidelity (of own data)


- Guided analysis
  - information scent - in complex data - scent about trends and relationships



- Presentation and communication
  - guidance - sequence
  - specificity/focus
  - semantics - explaining the context
  - efficiency - being mindful of attention spans
  - emphasis 
  - expressiveness - matching visuals with data properties; enables conveying the message
  - low visual complexity (time + attention)
  
 > <div class="alert alert-block alert-warning"> 
 In presentation/communication the point/message is defined by the author while in analysis is open-ended.
 </div>

### Criteria for evaluating explanatory visualisations 
Kosara et al. (2016)

- drawing interest (competing for attention)
- engagement
- willingness to explore (interactive visuals)
- memorability
- persuasion
- inspiring action

````python
######## EXAMPLE 1: APPLE STOCK PRICE OVER TIME ###############

### the data ####
df_app = pd.read_csv('apple_stock.csv')

df_app.reset_index(inplace=True) #reseting index to make "Date" a column (as it will be used in the plot)
df_app['Date'] = pd.to_datetime(df_app['Date'])
df_app = df_app.sort_values(by='Date')

#### BASIC PLOT ####

fig1 = px.line(df_app, x='Date', y='Close', title='Stock [closing] Price Over Time')
fig1.show() #to show the plot
````

````python
#### CUSTOMISATIONS ####

highlight_date = '2012-09-12'

fig1.update_traces(line=dict(color="blue", width=2))  # Change line color & thickness
fig1.update_layout(xaxis_rangeslider_visible=True)   # Add interactive range slider   
fig1.update_yaxes(title="Stock Price (USD)", showgrid=True)  # Customize y-axis
fig1.add_hline(y=df_app["Close"].mean(), line_dash="dash", line_color="red", annotation_text="Average Price")  # Add reference line for average stock price
fig1.add_trace(go.Scatter(x=[highlight_date], y=[702.1], marker=dict(color='blue', size=15, symbol='star'),
                         #text=[f'{highlight_date}: $702.1'],  # Display the date and value as text
                         textposition='top center',  # Position the text above the point
                         name='MAX. price',  
                         showlegend=False))
````

<div class="alert alert-block alert-success">
<b>IMPROVEMENTS FOR A DATA STORY:</b> 
</div>

- Better control of the timeline
- Comparison to overall average
- Clear indication/highlight of the highest value
- Interactivity

````python
###### EXAMPLE 2: AVERAGE GLOBAL TEMPERATURE LINE CHART #######

### the data ###
url_temp = "https://datahub.io/core/global-temp/r/annual.csv"
df_temp = pd.read_csv(url_temp)

##### BASIC PLOT #####
fig2 = px.line(df_temp, x="Year", y="Mean", title="Global Temperature Over Time")
fig2.show()
````

````python
#### EXAMPLE 3: HEATMAP FOR TIPS DATA ####

# we will use the tips dataset from the above, yet we first need to create crosstabulated frequencies
heatmap_data = pd.crosstab(tips['time'], tips['day'])
print(heatmap_data.head())

#### BASIC PLOT ####
fig8 = px.imshow(heatmap_data.values,
                labels=dict(x="Day", y="Meal", color="Count"),
                x=heatmap_data.columns,
                y=heatmap_data.index,
                title="Heatmap of Tips Dataset: Day vs. Meal")
fig8.show()
````

````python
#### CUSTOMISATIONS ####
fig8 = px.imshow(heatmap_data.values,
                labels=dict(x="Day", y="Meal", color="Count"),
                x=heatmap_data.columns,
                y=heatmap_data.index,
                color_continuous_scale='Reds', #color change
                title="Heatmap of Tips Dataset: Day vs. Meal")
fig8.update_traces(hovertemplate='Meal: %{y}<br>Day: %{x}<br>Count: %{z}<extra></extra>') #hoovering specification
#details: %{y}: Displays the value of the y variable with the label "Meal:";
#<br>: Inserts a line break in the hover label;
#Day: %{x}: Displays the value of the x variable with the label "Day:".
#Count: %{z}: Displays the value of the z variable with the label "Count:".
#<extra></extra>: Removes the trace name from the hover label, which is displayed by default. This is useful for decluttering the hover information.

#fitting the size of a plot to a screen
fig8.update_layout(
    autosize=True,
    margin=dict(l=0, r=0, t=0, b=0), #configures all margins (measures as pixels not mm)
    title_automargin=True  #  margin adjustment for the title
)

fig8.show()
````

<div class="alert alert-block alert-success">

### Proportions

````python
#### EXAMPLE 1: TREEMAP FOR GLOBAL PS4 GAMES SALES ####

#### the data ####
df_games = pd.read_csv('game_sales.csv', encoding='latin-1')
print(df_games.head())

# Group by genre and sum global sales
df_filtered_g = df_games.groupby("Genre")["Global"].sum().reset_index()

#### BASIC PLOT ####
fig8 = px.treemap(df_filtered_g, path=["Genre"], values="Global", 
                 title="Global Video Game Sales by Genre")
fig8.show()
````

````python
#### CUSTOMISATIONS ####

#adding percentages of sales
# Calculate percentages
df_filtered_g["Percentage"] = (df_filtered_g["Global"] / df_filtered_g["Global"].sum()) * 100
# Create formatted text labels (e.g., "Action: 25.3%")

fig8 = px.treemap(df_filtered_g, path=["Genre"], 
                 values="Global",
                 custom_data=["Percentage"],
                 title="Global Video Game Sales by Genre")

fig8.update_layout(title_font_size=20, title_x=0.5)  # Center title and adjust size

fig8.update_traces(
    textinfo="label+text",  # Show genre name + percentage
    texttemplate="%{label}<br>%{customdata[0]:.1f}%",  # Format text labels
    marker=dict(colorscale="Oranges"),  # Set color scheme
    textfont_size=16 #increase font size
)

fig8.show()
````

<div class="alert alert-block alert-success">
<b>IMPROVEMENTS FOR A DATA STORY:</b> 
</div>

- "Softer" color palette (single color)

````python
##### EXAMPLE 1: CONTINUOUS+CATEGORICAL BOXPLOT - MILES PER GALLON ACROSS CAR TYPES ######

#### THE DATA #####

df = autompg2[["class", "hwy"]].rename(columns={"class": "kind"})

kinds = df.kind.unique()

qs = df.groupby("kind").hwy.quantile([0.25, 0.5, 0.75])
qs = qs.unstack().reset_index()
qs.columns = ["kind", "q1", "q2", "q3"]
df = pd.merge(df, qs, on="kind", how="left")

# compute IQR outlier bounds
iqr = df.q3 - df.q1
df["upper"] = df.q3 + 1.5*iqr
df["lower"] = df.q1 - 1.5*iqr

source = ColumnDataSource(df)

#### BASIC PLOT ####
p3 = figure(x_range=kinds, tools="", toolbar_location=None,
           title="Highway MPG distribution by vehicle class",
           background_fill_color="#eaefef", y_axis_label="MPG")

# outlier range
whisker = Whisker(base="kind", upper="upper", lower="lower", source=source)
whisker.upper_head.size = whisker.lower_head.size = 20
p3.add_layout(whisker)

# quantile boxes
cmap = factor_cmap("kind", "TolRainbow7", kinds)
p3.vbar("kind", 0.7, "q2", "q3", source=source, color=cmap, line_color="black")
p3.vbar("kind", 0.7, "q1", "q2", source=source, color=cmap, line_color="black")

# outliers
outliers = df[~df.hwy.between(df.lower, df.upper)]
p3.scatter("kind", "hwy", source=outliers, size=6, color="black", alpha=0.3)

p3.xgrid.grid_line_color = None
p3.axis.major_label_text_font_size="14px"
p3.axis.axis_label_text_font_size="12px"

show(p3)
````

````python
#### CUSTOMISATIONS (AND CODE EXPLANATION) ####
import altair as alt
from vega_datasets import data

source = data.unemployment_across_industries.url

selection = alt.selection_point(fields=['series'], bind='legend') 
#alt.selection_point(): Creates an interactive selection.
#fields=['series']: Specifies that we want to filter by the series field/variable (industry type).
#bind='legend': Links the selection to the chart legend

# Create the chart
chart = alt.Chart(source).mark_area().encode(
    alt.X('yearmonth(date):T', axis=alt.Axis(domain=False, format='%Y', tickSize=0)), 
    #yearmonth(date):T: Converts the date to year-month format (T means time)
    #domain=False: Removes the axis line.
    #format='%Y': Displays only the year.
    #tickSize=0: Hides tick marks.
    alt.Y('sum(count):Q', stack='center', axis=None),
    #sum(count): Sums up the count of unemployment cases.
    #:Q (quantitative): Indicates that this is numerical data.
    #stack='center': Uses center stacking (good for area charts).
    #axis=None: Hides the y-axis labels.
    alt.Color('series:N', scale=alt.Scale(scheme='tableau10')),  # Custom color scheme
    #:N (nominal) means categorical data (e.g., industry names).
    opacity=alt.when(selection).then(alt.value(1)).otherwise(alt.value(0.2)), 
    # 1 - fully viisble; 0.2 - faded
    tooltip=['yearmonth(date):T', 'series:N', 'sum(count):Q']  # Tooltip for interactivity
).add_params(
    selection
    #adding interactive legend
).properties(
    title="Unemployment Across Industries Over Time"  # adding title
)

chart
````

````python
#### CUSTOMISATIONS ####

alt.theme.enable("fivethirtyeight")  #changing the theme

#built-in themes in Altair:
    #quartz
    #excel
    #ggplot2
    #fivethirtyeight
    #dark
    #latimes
    #googlecharts
    #vox
    #urbaninstitute
    #powerbi
    #carbong10
    #carbong90
    #carbong100
    #carbonwhite

highlight = alt.selection_point(on='pointerover',
                          fields=['symbol'], nearest=True)
#alt.selection_point() Creates an interactive selection that reacts to user input.
#on='pointerover' Triggers the selection when the mouse hovers over a data point.
#fields=['symbol'] Highlights elements based on the symbol field (each stock's identifier).
#nearest=True Ensures the selection highlights the nearest data point to the pointer.

base = alt.Chart(source1).encode(
    x='date:T',
    y='price:Q',
    color=alt.Color('symbol:N', scale=alt.Scale(range=['red', 'blue', 'green', 'purple', 'orange'])) 
)

#x='date:T' Maps the date column (T means temporal/time data) to the x-axis.
#y='price:Q' Maps the price column (Q means quantitative/numeric data) to the y-axis.
#color='...' Assigns specified range of colors to symbol variable (N means nominal/categorical data).

points = base.mark_circle().encode(
    opacity=alt.value(0)
).add_params(
    highlight
).properties(
    width=700,  # Adjusted width
    height=400, # Adjusted height
)

#mark_circle()  Adds circular markers to the chart.
#encode(opacity=alt.value(0))  Makes the circles invisible (opacity=0).
#add_params(highlight) Links the highlight interaction to these invisible circles.
#properties(width=700, height=400)  Sets the chart size.

lines = base.mark_line().encode(
    size=alt.condition(highlight, alt.value(4), alt.value(1)) 
)

#mark_line() Draws lines connecting the data points.
#encode(...) Defines the line's properties:
#size=alt.condition(highlight, alt.value(4), alt.value(1)) Uses alt.condition() to change the line thickness dynamically.
#highlight is the condition (if the symbol is hovered).
    #If highlight is active Line size is 4.
    #If highlight is not active Line size is 1.

points + lines #overlays the marks on the same chart.
````

### Brief overview of ML

- Quantitative modelling cultures
  - DATA MODELLING – parametric/conventional; fitting the parameters to a model by using data; explanative objective; deductive
  - **ALGORITHMIC MODELLING – ML** -  finding a function (an algorithm) that operates on X to predict Y; predictive objective; testing predictions on unseen data; "inductive"

- What is ML? 
  - Learning from (training) data and applying this knowledge on yet unseen (test/new) data &rarr; **prediction**
  - Applications: quant. text analysis, climate models, crime rate prediction, education outcomes, etc.

- > **Supervised ML**: Defining the target variable; classification task if dependent/target variable is categorical, regression task if target variable is continuous

- > **Unsupervised ML**: target variable not defined; uncovering a pattern within a dataset or to reduce dimensionality in our data

- Advantages:
   - utilising data abundance
   - combining different data tpyes
   - "light" modelling assumptions
   
- Weaknesses
  - model interpretation?
  - context?
  - model selection?

#### Before we dvelve into the example, a brief overview of decision tree model...



##### Basic decision tree algorithm (ID3; C4.5)
- Start splitting Xs in the training set - Which Attribute Is the Best Classifier?
- In order to answer this question, we calculate a statistical property (e.g., Entropy, Information gain, etc.) of each X. 
- The X that yields the best result is the root node (i.e., the 1st split)


- The splitting continues until:
  - (1) (theoretically) every attribute has already been included along this path (conjunction) through the tree;
  - (2) The training examples associated with this leaf node all have the same target attribute value (with this we are likely to overfit the model)

##### Basic decision tree algorithm (ID3; C4.5)

- Vital question: WHEN DO WE STOP SPLITTING? (to avoid overfit)
  - max_depth (max. number of splits in the model)
  - min_samples_split (split if the node contains certain number of examples/observations)
  - cross validation
  - tree pruning (to improve accuracy of the model for the test set)

- Measures for prediction evaluation - confusion matrix

````python
# loading packages 
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
````

````python
#getting (synthetic) data on car buying preferences
df_car = pd.read_csv('synthetic_car_preferences.csv')
df_car.head()
````

````python
# Split dataset into features and target variable
X = df_car.drop(columns=['Preferred_Car_Type'])
y = df_car['Preferred_Car_Type']

# Split data training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=5, random_state=42)
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Classification Report:\n', classification_report(y_test, y_pred))

# Visualize the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()

# Feature Importance Analysis
feature_importances = pd.Series(clf.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar', title='Feature Importances')
plt.show()
````

````python
#making features importance plot a bit nicer with plotly...
import plotly.express as px
feature_importances = pd.Series(clf.feature_importances_, index=X.columns)
fig_ml = px.bar(feature_importances.sort_values(ascending=True), color_discrete_sequence=['black'],
             title='Lifestyle factors main predictors of car types/designs!', orientation='h', labels={'index': 'Feature', 'value': 'Importance'})
fig_ml.update_layout(template="simple_white")
fig_ml.update_layout(showlegend=False)
fig_ml.show()
````

### Example 2: unsupervised ML - K-means algorithm: car CO2 emissions and fuel efficiency

#### Brief overivew of K-means clustering

- N.B. Before running the algorithm we need to specify the number of clusters **(k)** ex ante.
- PROS: simple to implement; results are intuitively presented; appropriate for large datasets; low computation cost
- CONS: difficult to determine optimal number of clusters; lack of consistency (due to random selection of centroids); outliers significantly affect the clustering process


- Algorithm:
   - (1) The number **(k)** of pre-determined clusters = the number of initial CENTROIDS  Centroid is RANDOMLY chosen from our data – it represents (real or “imaginary”) center of each cluster **k**
  - (2) We assign each data point to one of **k** clusters by calculating each data point’s distance from centroids
  - (3) We calculate new centroids = the mean of values in each cluster from previous step
  - (4) repeat (2) and (3) until centroids stop changing

In this example we are dettecting patterns (as groups/clusters) from a *(synthetic)* car dataset that includes 500 car models and 5 variables regarding their fuel efficiency and CO2 emissions. For the purposes of learning, as above, the data is sythetic. We wish to write a data story on this topic.

````python
df_car_effi = pd.read_csv('synthetic_car_CO2.csv')
df_car_effi.head()
````

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_car_effi)

# Fit KMeans model
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df_scaled)

# Add the cluster labels to the DataFrame
df_car_effi['Cluster'] = kmeans.labels_
````
